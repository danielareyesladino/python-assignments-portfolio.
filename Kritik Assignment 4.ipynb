```python
# Kritik Assignment 4

import numpy as np
import matplotlib.pyplot as plt    # import libraries

# code from assignment
def gradient_descent(f,learning_rate, initial_point):  
    def deriv (f,base_point):    # estimates derivative of function at base_point
        return (f(base_point+10**(-5))-f(base_point-10**(-5)))/(2*10**(-5))
    x_coords=[initial_point]   # List where x points are stored to make cobweb
    y_coords=[f(initial_point)] #List where y points are stored to make cobweb

    # defining the derivative of the function
    def function_derivative(x):
        return (f(x+0.00001)-f(x))/0.00001
    i=1  # counter initilizer for while loop 

    # loop that will bounce back from point to point until finding the min/max
    # 0.00001 is the margin of error
    while abs(function_derivative(x_coords[i-1])) >0.00001:
        xi= x_coords[i-1]- learning_rate*function_derivative(x_coords[i-1])
        x_coords.append(xi)   # storing each x point in the list
        y_coords.append(f(x_coords[i]))  # storing each y point in the list
        i=i+1

    # code from the assignment
    plot_range=np.linspace(min(x_coords) - 0.5, max(x_coords) + 0.5,1000)
    function_range=[f(i) for i in plot_range]
    plt.plot(plot_range, function_range, 'cornflowerblue')
    plt.scatter (x_coords, y_coords, color = 'maroon')
    plt.plot (x_coords, y_coords, color = 'maroon') 

    return round(x_coords[-1],3), round (y_coords[-1],3)

# graphing x^2
def function_test(x):
    return x**2

gradient_descent(function_test,0.9,1)

```
