```python
# Kritik 6: Finding the probability that the height of a randomely chosen man is between 162-190 cm

# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad  # new library that allows for integration, instead of having to use numpy 

# Part of code given in assignment to define function of normal density
def normal_density(mean, variance, x):
    exp_term = np.exp(-((x - mean)**2) / (2 * variance))   # the exponential fuction allows for bell curve shape
    return exp_term / np.sqrt(2 * np.pi * variance)

# Part of the code given for numerical integration of the normal density function
def integration(mean, variance, a, b):
    # Define the integrand (funtion being integrated, i.e. normal density function) using the normal density function
    integrand = lambda x: normal_density(mean, variance, x)
    
    # Use the quad function from scipy library to integrate 
    result,_ = quad(integrand, a, b)
    
    return result

# Plotting normal_density function for different values of  and ^2
x_values = np.linspace(150, 190, 100)

# for loop to find the probability (y values in graph). This also allows to have a couple different values of the mean and variance
# Hence why the final plot will have three bell curves
for mean, variance in [(171, 7.1**2), (170, 7**2), (175, 8**2)]: # three different mean and variance values that will be plotted on the same graph
    y_values = normal_density(mean, variance, x_values)
    plt.plot(x_values, y_values, label= f'mean={mean}, variance={variance}')


plt.title('Normal Density Function')
plt.xlabel('Height (cm)')
plt.ylabel('Probability')
plt.legend()
plt.show()

# calculating probability
mean_given = 171
variance_given = 7.1**2

probability = integration(mean_given, variance_given, a, b)
print(f"The probability that a randomly chosen man has a height between 162 cm and 190 cm is: {probability:.4f}")
```
