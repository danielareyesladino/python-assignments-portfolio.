```python
# Kritik Assignment 9

import sympy as sp
import numpy as np

# a)
# define x,y and function
x,y = sp.symbols('x y')
f = sp.exp(x)*sp.sin(y) + sp.sin(y**3)

# use sympy for partial derivatives
df_dx = sp.diff (f,x)
df_dy = sp.diff (f,y)

# print answers
print (df_dx)
print (df_dy)
```

```python
#b)
import sympy as sp

# define x,y and function
r,t = sp.symbols ('r t') 
f = (r**2)*t + r*t**2

#Find partial derivatives
df_dr = sp.diff (f, r)
df_dt = sp.diff (f, t)

#Plug in (1, -1) in partial derivatives to find magniture of gradient
# Use subs documentation to be able to find partial derivative at a specific point
point = [(r, 1), (t, -1)]
df_dr_value = df_dr.subs(point)
df_dt_value = df_dt.subs(point)

gradient_vector = [df_dr_value, df_dt_value]

magnitude = sp.sqrt(df_dr_value**2 + df_dt_value**2)

print(gradient_vector)
print(magnitude.evalf())
```

```python
#c)

# define x,y and function
a,b = sp.symbols('a b')
g = sp.log(a**2 + b**2)

#Find partial derivatives
dg_da = sp.diff (g,a)
dg_db = sp.diff (g,b)

#Find second partial derivatives
d2g_da2 = sp.diff (dg_da,a)
d2g_db2 = sp.diff (dg_db,b)

#Find mixed derivative
d2g_dadb = sp.diff (dg_db,a)

print (d2g_da2)
print (d2g_db2)
print (d2g_dadb)
```

```python
#d)
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
#e)
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
j = sp.sin(x) +sp.cos(y)

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-20, 10, 400)
y_vals = np.linspace(-20, 10, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='plasma')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = sin(x) + cos(y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
